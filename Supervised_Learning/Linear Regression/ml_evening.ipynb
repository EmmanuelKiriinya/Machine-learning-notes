### 1. Linear Regression.  
Linear regression is a supervised machine learning algorithm and statistical method used to model the relationship between a dependent variable(target) and one or more independent(predictor) variables.     
Its primary objectives are to:     
- Predict a continous value.     
- Understand how predictors(independent variables) influence the target(dependent variable)  
- Quantifying relationships between variables.    

_Types of Linear Regression_   
1. Simple linear regression => Models the relationship between a single dependent variable and one independent variable using a straight line.     
2. Multiple linear regression => Models the relationship between a single dependent variable and two or more independent variables.   
3. Polynomial linear regression => Models a non-linear relationship by fitting a curved line throught the data using polynomial terms. e.g $y = {0}_1 {x} + {0_2}x^2 + ... {0_n}x^n$    

#### 1.1 Simple Linear Regression.  
Uses one predictor variable(independent variable).  
_Example_    
Predicting salary using years of experience.    

_Formula_   
$y = mx + c$  

ŷ =  β₁x + β₀

y = predicted value or the dependent variable.   
m or β₁ = Slope or the gradient(represents the expected change in Y for every 1-unit increase in X).   
x = predictor or the independent variable.   
c or β₀ = y-Intercept(this is te value of y when X is equal to zero).      

_Formula for slope_    
${beta_1} = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n}(x_i - \bar{x})^2}$ 

alternative formula:    
$m=\frac{n(\sum xy)-(\sum x)(\sum y)}{n(\sum x^{2})-(\sum x)^{2}}$

_Formula for the intercept_    
$β₀ = ȳ - β₁x̄$  

_Example_  
|Experience(x)|Salary_Kes(y)|
|-------------|-------------|
|1.3          |42500        |
|3.2          |63500        |
|7.0          |81800        |
|9.4          |89500        |


````python
import pandas as pd

df = pd.DataFrame({'experience':[1.3,    3.2,    7.0,    9.4   ],
                   'salary':[42500, 63500, 81800, 89500]})

# Get the squares of x
df["experience_squared"] = df["experience"]**2

# Multiplt x and y
df["xy"] = df["experience"] * df["salary"]

# SLope
slope = ((len(df)*(df["xy"].sum())) - (df["experience"].sum() * df["salary"].sum()))/(((len(df)*df["experience_squared"].sum()) - ((df["experience"].sum())**2)))

# Intercept9
intercept = (df["salary"].mean()) - (slope * (df["experience"].mean()))

print("Slope", slope)
print("Y-Intercept", intercept)
````

- __Slope:__ The slope tells us how much Y changes when X increases by one unit.

#### Assumptions of Linear Regression.  
1. __Linearity:__ The relationship between predictors and the target should be approximately linear. To check for linearity you use a scatterplot.    

![Linear Scatterplots](https://i.ytimg.com/vi/44f5YOI7E1k/sddefault.jpg)    

2. __Independence of errors:__ Residuals should be independent of one another.   
3. __No Multicollinearity:__ Predictors/independent variables should not be highly correlated with each other. You can use correlation matrix, Variance Inflation Factor(VIF).   
_VIF Formula_     
$VIF = \frac{1}{1 - R_j^2}$    

Interpretation:      
VIF = 1: No multicollinearity, the variables are completely independent.     
VIF 1 to 5: Moderate correlation; Usually not an issue for regression models.     
VIF 5 to 10: High Correlation amongst independent variables.    
VIF > 10: Serious multicollinearity.    

4. __No influential outliers:__ Extreme outliers/observations can disproportionately affect the regression line.    

5. __Normality of residuals:__ Residuals should be approximately normally distributed, especially for valid confidence intervals and hypothesis tests.

## Model Evaluation metrics.    
Model evaluation metrics compare the model's prediction and predicted values with the actual observed values and quantify the prediction error. The metrics help us answer questions such as:     
- How accurate are the predictions?    
- Which regression model performs better?    
- Are the prediction errors small enough to be acceptable?
- Does the model explain a large proportion of the variation in the target variable?     

No single metric is sufficient on its own. In practice, it is common to examine several evaluation metrics together.        

### 1. Mean Absolute Error(MAE)    
The MAE measures the average magnitude of the prediction errors without considering whether the model overestimates or underestimates the target variable.      

_Formula_      
$MAE=\frac{1}{n}\sum_{i=1}^{n}|y_i-\hat{y}_i|$       

$y_i$ = Actual value.    
$\hat{y}_i$ = Predicted value.   
n = number of observations/rows.      

_Example_   
Suppose we are predictin house prices.      

|Actual prices|predicted prices|absolute error|
|-------------|----------------|--------------|
|250000       |245000          |5000          |
|300000       |290000          |10000         |
|395000       |400000          |5000          |     

The average absolute error is:    



````python
total_abs_errors = 5000 + 10000 + 5000
print("Total absolute errors: ", total_abs_errors)

MAE = total_abs_errors/3

print("Mean Absolute Error: ", MAE)
````

This means that, on average, the model's predictions are off by __6667__ Ksh.    

_Advantages_      
- Very to understand.   
- Expressed in the same units as the target variable.   
- Less sensitive to to extreme errors than MSE and RMSE.    
- Suitable when all prediction errors should be treated equally.  

_Disadvantages_     
- Does not penalize large errors more heavily than small ones.    
- Two models with same MAE may have different error distributions.    

### 2. Mean Squared Error(MSE)
MSE Calculates the average of the squared prediction errors.     

_Formula_    
$MSE=\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y}_i)^2$


Because the errors are squared:    
- Larger errors contribute much more than smaller ones.   
- Negative errors become positive.  
- Models that make a few large mistakes recieves a higher penalty.     

_Example_.    

|Actual|Predicted|error/residual|squared errors|
|------|---------|--------------|--------------|
|20    |22       |-2            |4             |
|39    |36       |3             |9             |
|30    |40       |-10           |100           |    

Notice that the error of 10 contributes 100 to the MSE, makes it more influential than the smaller errors.       

````python
total_squared_errors = 4 + 9 + 100
MSE = total_squared_errors/3

print("Sum of Squares: ", total_squared_errors)
print("Mean Squared Error", MSE)
````

_Advantages_     
- Penalizes large prediction errors heavily.    
- Forms the basis of Ordinary Least Squares(OLS)
- Good for optimization task in mathematics.    

_Disadvantages_    
- Difficult to interpret because the units are squared.   
- Highly sensitive to outliers.    

### 3. Root Mean Squared Error(RMSE).   
The RMSE is the square root of the MSE.     

$RMSE = \sqrt{MSE}$ 

Taking the square root converts the error back to the original units of the target variable.      

_Interpretation_    
MSE = 37.67

Then the RMSE = $\sqrt{37.67}$ = 6.14    

The model's predictions differ from the actual values by approximately 6.14 units.      

_Advantages_    
- Easy to interpret because it has the units as the target variable.  
- Penalizes larger prediction errors.  

_Disadvantages_     
- Sensitive to outliers.   
- Can be heavily influenced by a few prediction errors.    




````python
import numpy as np
np.sqrt(37.67)
````

### 4. R<sup>2</sup> (Coefficient of determination).   
The R<sup>2</sup> Score measures how well the regression model expalains the variation in the depenednt variable.    

_Formula_      
$R^2=1-\frac{SS_{Residual}}{SS_{Total}}$ 

$SS_{Residual}$ = Sum of squared Residuals(Unexplained variation).   
$SS_{\text{res}}=\sum _{i=1}^{n}(y_{i}-\^{y}_{i})^{2}$    

$SS_{Total}$ = Total Sum of Squares (Total variation in the target variable)  

$SS_{\text{tot}}=\sum_{i=1}^{n}(y_{i}-\={y})^{2}$ 

$y_{i}$ = Actual values.   
$\={y}$ = Mean average of the observed values.   
n = total number of observations/number of rows    

_Interpretation_     
|R<sup>2</sup> Value|Interpretation                                                 |
|-------------------|---------------------------------------------------------------|
|0.00               |The model explains none of the variation in the target variable|
|0.40               |The model explains 40% of the variation                        |
|0.75               |The model explains 75% of the variation                        |
|1.00               |The model explains 100% of the variation/perfect prediction    |  

A R2 score of 0.8 means that 80% of variation in the target variable is explained by the predictors, while remaining 20% is due to factors not captured by the model or random variation.      

_Advantages_    
- Easy to interpret as a percentage of explained variation.   
- Useful for comparing models built on the same dataset.   
- Provides overall measure of the model fit.    

_Disadvantages_    
- A high R2 does not necessarily mean the model is appropriate.    
- Adding more predictors can artificially increase R2, even if these predictors have little predictive value.    
- Does not indicate whether the regression assumptions have been met.   

### Adjusted R<sup>2</sup> Score.   
The Adjusted R2 improves upon R2 by accounting for the number of predictors in the model.    
Unlike R2, Adjusted R2 increases only when a new predictor genuinely improves the model's performance. If an irrelevant predictor is added, Adjusted R2 may remain unchanged or even decrease.     

_Example_     
Suppose we have two models predicting house prices:    
- __Model A:__ Size, Bedrooms   
- __Model B:__ Size, Bedrooms, House Color.  

House color is unlikely to influence house prices. Although adding it may slightly increase the R2 value, it provides no meaningful predictive power. Adjusted R2 helps identify this by penalizing unnecessary predictors.    

_Adjusted R2 Advantages_    
- Reduces the risk of overfitting.   
- Encourages selecting only meaningful predictors.    

_Disadvantages_     
- Slightly more difficult to interpret than R2
- Still should not be used as the sole criterion for the model selection.
