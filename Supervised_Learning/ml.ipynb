## Introduction to Machine Learning & Types of Learning
### 1. What is Machine Learning?

Machine Learning is a branch of Artificial Intelligence that allows computers to learn patterns from data and make predictions or decisions without being explicitly programmed for every rule.  
Instead of writing fixed instructions, we give the computer data, and it learns relationships from that data.   
For example, instead of manually writing rules to determine house rent in Nairobi, we can train a machine learning model using previous housing data. The model can learn how factors such as estate, number of bedrooms, house size, availability of parking, and distance to the CBD influence rent prices.  

__Simple Example__  

Assume we have housing data with the following columns:  
|Estate	| Bedrooms |	House Size |	Parking |	Monthly Rent |
|-------|----------|---------------|------------|----------------|
|Umoja	|2	       |80 sqm	       |No	        |45,000          |
|Karen	| 4	       |220 sqm	       |Yes	        |250,000         |
|Kilimani|3	       |150 sqm	       | Yes	    | 120,000

A machine learning model can study this data and learn that:  
- Houses in high-end estates usually have higher rent.
- More bedrooms often increase rent.
- Bigger house size may lead to higher rent.
- Parking availability can affect price.  

After learning from past data, the model can predict rent for a new house.

## 4. Types of Machine Learning
Machine learning can be grouped into three major types:  
- Supervised Learning
- Unsupervised Learning
- Reinforcement Learning

### 4.1 Supervised Learning

Supervised learning is a type of machine learning where the model learns from labelled data.  
Labelled data means that the dataset already contains the correct answer.    
_Example_   
If we are predicting monthly rent, the column monthly_rent_kes is the label because it is the answer we want the model to learn.  

| Estate |	Bedrooms |	House Size(sqm)|Parking|Monthly Rent|
|--------|-----------|-----------------|-------|------------|
|Umoja	 |2	         |80               |No     |45,000      |
|Karen	 |4	         |220 	           |Yes	   |250,000     |
|Kilimani|3	         |150              |Yes	   |120,000     |

Here:  
__Features:__ Estate, Bedrooms, House Size, Parking. 
__Label:__ Monthly Rent.  
The model learns the relationship between the features and the label.  
Common Supervised Learning Tasks.  

__a) Regression__   
Regression is used when the target variable is numeric.    
Examples:   
Predicting house rent.  
Predicting salary.  
Predicting sales revenue.   
Predicting student exam score.      

__b) Classification__    
Classification is used when the target variable is a category.   
Examples:  
Predict whether a customer will default: Yes or No.  
Predict whether an email is spam or not spam.  
Predict whether a student will pass or fail.  
Predict whether a house is affordable or expensive.  

## 4.2 Unsupervised Learning. 
Unsupervised learning is used when the dataset does not have a labelled answer.  
The model tries to discover hidden patterns, groups, or structures in the data by itself.  

_Example_      
Assume we have housing data with estate, rent, house size, bedrooms, and distance to CBD, but we do not give the model a target column to predict.     
The model may group houses into clusters such as:    
Budget houses     
Middle-income houses     
Luxury houses.   

___Common Unsupervised Learning Tasks___      
__a) Clustering__  
Clustering means grouping similar records together.  
Example:  
Group Nairobi houses into different market segments based on rent, bedrooms, house size, and location.     
Possible clusters:  

Cluster	Description   
Cluster 1	Low-cost rental houses.    
Cluster 2	Middle-income apartments.  
Cluster 3	High-end luxury houses.   

__b) Dimensionality Reduction.__     
Dimensionality reduction reduces many columns into fewer important components while still preserving useful information.   
It is useful when a dataset has many columns.   

Examples:  
Reducing many customer behavior variables into a few customer profiles.     
Reducing many housing features into fewer property quality indicators.  

## 4.3 Reinforcement Learning.  
Reinforcement learning is a type of machine learning where an agent learns by interacting with an environment.   
The agent takes actions and receives rewards or penalties.   
The goal is to learn the best actions that maximize reward over time.   

_Simple Example_    

Think of a robot learning how to walk.   
If it moves correctly, it gets a reward.   
If it falls, it gets a penalty.   
Over time, it learns the best way to walk.   
Real-World Examples.  

Reinforcement learning is used in:  

Robotics. 
Self-driving cars. 
Game-playing AI. 
Automated trading. 
Route optimization. 
Dynamic pricing. 

_Example:_    
A delivery company can use reinforcement learning to find the best route that reduces fuel cost, delivery time, and traffic delays.    

## 5. Concept of Datasets in Machine Learning.  

A dataset is a collection of data used to train, test, and evaluate a machine learning model.  
In most machine learning tasks, data is organized in rows and columns.  
Rows represent observations or records.  
Columns represent variables or attributes.  

_Example:_  

| Estate |	Bedrooms |	House Size(sqm)|Parking|Monthly Rent|
|--------|-----------|-----------------|-------|------------|
|Umoja	 |2	         |80               |No     |45,000      |
|Karen	 |4	         |220 	           |Yes	   |250,000     |
|Kilimani|3	         |150              |Yes	   |120,000     |   

Each row represents one property.   
Each column gives information about that property.  

## 6. Features and Labels 
### 6.1 Features  
Features are the input variables used by the model to make predictions.   
In a housing dataset, features may include:   
Estate.  
Bedrooms.  
Bathrooms.  
House size.   
Parking availability.     
Distance to CBD.    
Property type.   
Security availability.    

These are the columns the model uses to learn patterns.    

### 6.2 Label   
A label is the target variable or the answer we want the model to predict.  
In a rent prediction problem, the label is:  
Monthly Rent.   
In a loan approval problem, the label could be:   
Approved or Not Approved.  
In a student performance problem, the label could be:   
Pass or Fail.    

Example:    
Feature Columns	Label Column.    
Estate, Bedrooms, Bathrooms, House Size, Parking	Monthly Rent.   

## 7. Training Data and Testing Data.    
When building a machine learning model, we usually divide the dataset into two main parts:    
Training data.  
Testing data.  


### 7.1 Training Data.   
Training data is the portion of the dataset used to teach the model.    
The model studies this data and learns patterns from it.    

_Example:_  
The model learns that houses in Karen with more bedrooms and larger house sizes usually have higher rent.    

### 7.2 Testing Data.   
Testing data is used to evaluate how well the model performs on new, unseen data.    
This helps us know whether the model has truly learned patterns or has simply memorized the training data.    

7.3 Example Split

If we have 1,000 housing records, we can split them as follows:

Dataset Part	Percentage	Number of Records
Training data	80%	800
Testing data	20%	200

The model learns from the 800 records and is tested on the remaining 200 records.

9. Summary

Machine learning allows computers to learn from data and make predictions or decisions.

There are three major types of machine learning:

Type	Description	Example
Supervised Learning	Learns from labelled data	Predict house rent
Unsupervised Learning	Finds hidden patterns without labels	Group houses into market segments
Reinforcement Learning	Learns through rewards and penalties	Self-driving car learning best routes

Important dataset concepts include:

Concept	Meaning
Dataset	Collection of data
Feature	Input column used for prediction
Label	Target column to be predicted
Training data	Data used to teach the model
Testing data	Data used to evaluate the model

In simple terms:

Features are the questions, the label is the answer, training data is used for learning, and testing data checks whether the model has learned well.
