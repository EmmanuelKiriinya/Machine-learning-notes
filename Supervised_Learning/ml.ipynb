## Introduction to Machine Learning & Types of Learning
### 1. What is Machine Learning?

Machine Learning is a branch of Artificial Intelligence that allows computers to learn patterns from data and make predictions or decisions without being explicitly programmed for every rule.  
Instead of writing fixed instructions, we give the computer data, and it learns relationships from that data.   
For example, instead of manually writing rules to determine house rent in Nairobi, we can train a machine learning model using previous housing data. The model can learn how factors such as estate, number of bedrooms, house size, availability of parking, and distance to the CBD influence rent prices.

# 1. Supervised learning  
## 1.1 Linear Regression 
Linear regression is a supervised machine learning algorithm and statistical method used to model the relationship between a dependent (target) variable and one or more independent (predictor) variables.   
Its primary objectives are to:  
- Predict a continuous value
- Understand how predictors influence the target
- Quantify relationships between variables 

_Types of linear regression_    
a). Simple Linear regression.  
b). Multiple Linear regression.  

___Assumptions of Linear Regression___    
These assumptions are essential because violations can lead to biased estimates, incorrect statistical tests, or poor predictions.   
1. __Linearity:__ The relationship between predictors and the target should be approximately linear.  
    How to check: Scatter plots or residuals vs. fitted values.   
2. __Independence of Errors:__ Residuals should be independent of one another.   
This is particularly important for time series data.

3. __Homoscedasticity (Constant Variance):__ Residuals should have roughly the same spread across all fitted values.  
    If the spread increases or decreases systematically, the model exhibits heteroscedasticity.    

4. __Normality of Residuals:__ Residuals should be approximately normally distributed, especially for valid confidence intervals and hypothesis tests.   

5. __No Multicollinearity (Multiple Regression):__ Predictors should not be highly correlated with each other. A common diagnostic is the Variance Inflation Factor (VIF).  
6. __No Influential Outliers:__ Extreme observations can disproportionately affect the regression line.
### 1.1.1 Simple Linear Regression.  
Uses one predictor variable.   
_Example_  
Predict salary using years of experience.   

Equation:

y=β0 + β1x or y=mx + c  

y = predicted value  
β₀ = intercept  
β₁ = slope  
x = predictor  

### 1.1.2 Multiple Linear Regression  
Uses two or more predictors.    
_Example_   
Predict house price using  
Size   
Bedrooms   
Age   
Location score    

_Equation_  
y=β0+β1x1+β2x2+⋯+βnxn    

__Dependent vs Independent Variables__    
_Dependent_   
Independent Variable (X)  
The predictor.  
E.g
House size  
Advertising budget  
Age  
Temperature  

Dependent Variable (Y)  
The outcome being predicted.  
E.g  
House price  
Sales  
Medical expenses  
Electricity consumption   

#### Definition of terms in Linear regression.    
- __The Regression Line (Best Fit Line):__ The regression line summarizes the average relationship between X and Y. Each point has a predicted value that lies on the line. The line minimizes prediction errors.    
- __Intercept (β₀):__ The intercept is the predicted value of Y when X = 0.  
- __Slope (β₁):__ The slope tells us how much Y changes when X increases by one unit.  
    _Example_  
    Sales=100+8×Advertising  
    ___Interpretation___
    Every additional $1 spent on advertising increases sales by 8 units on average.  

- __Residuals (Errors):__ Residual = Actual value − Predicted value  
Residual=$y - \hat{y}$    
y = The actual figure     
$\hat{y}$ = The predicted figure 

    Residuals measure prediction error.

    Positive residual => Model underestimated.    
    Negative residual => Model overestimated.    
  
- __Least Squares Method:__ The model chooses the regression line that minimizes the sum of squared residuals.   
    Instead of minimizing(which cancels positive and negative errors):      
    $\sum(y - \hat{y})$    
    it minimizes:   
    $\sum(y - \hat{y})^{2}$   

    This is called the __Ordinary Least Squares (OLS)__ method.



## 13. Model Evaluation Metrics

After building a linear regression model, we need to determine **how well it performs**. Model evaluation metrics compare the model's predicted values with the actual observed values and quantify the prediction error. These metrics help us answer questions such as:

* How accurate are the predictions?
* Which regression model performs better?
* Are the prediction errors small enough to be acceptable?
* Does the model explain a large proportion of the variation in the target variable?

No single metric is sufficient on its own. In practice, it is common to examine several evaluation metrics together.

---

### 13.1 Mean Absolute Error (MAE)

The **Mean Absolute Error (MAE)** measures the average magnitude of the prediction errors without considering whether the model overestimates or underestimates the target variable.

The formula is:

$MAE=\frac{1}{n}\sum_{i=1}^{n}|y_i-\hat{y}_i|$

where:

* $y_i$ = Actual value
* $\hat{y}_i$ = Predicted value
* (n) = Number of observations

#### Interpretation

Suppose we are predicting house prices.

| Actual Price | Predicted Price | Absolute Error |
| ------------ | --------------- | -------------- |
| 250,000      | 245,000         | 5,000          |
| 300,000      | 290,000         | 10,000         |
| 400,000      | 395,000         | 5,000          |

The average absolute error is:


$MAE=\frac{5000+10000+5000}{3}=6,667$

This means that, on average, the model's predictions are off by **6,667 units**.

#### Advantages

* Very easy to understand.
* Expressed in the same units as the target variable.
* Less sensitive to extreme errors than MSE and RMSE.
* Suitable when all prediction errors should be treated equally.

#### Disadvantages

* Does not penalize large errors more heavily than small ones.
* Two models with the same MAE may have very different error distributions.


### 13.2 Mean Squared Error (MSE)

The **Mean Squared Error (MSE)** calculates the average of the squared prediction errors.

The formula is:

$MSE=\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y}_i)^2$

Because the errors are squared:

* Larger errors contribute much more than smaller ones.
* Negative errors become positive.
* Models that make a few very large mistakes receive a higher penalty.

#### Example

Prediction errors:

* 2
* 3
* 10

Squared errors:

* 4
* 9
* 100

Notice that the error of 10 contributes **100** to the MSE, making it much more influential than the smaller errors.

#### Advantages

* Penalizes large prediction errors heavily.
* Widely used in optimization because it is mathematically differentiable.
* Forms the basis of Ordinary Least Squares (OLS) regression.

#### Disadvantages

* Difficult to interpret because the units are squared.
* Highly sensitive to outliers.

---

### 13.3 Root Mean Squared Error (RMSE)

The **Root Mean Squared Error (RMSE)** is the square root of the Mean Squared Error.

The formula is:

$RMSE=\sqrt{MSE}$  

Taking the square root converts the error back to the original units of the target variable.

#### Interpretation

Suppose:

MSE = 64

Then:


$RMSE=\sqrt{64}=8$  

This means that the model's predictions differ from the actual values by approximately **8 units on average**.

#### Advantages

* Easy to interpret because it has the same units as the target variable.
* Penalizes large prediction errors.
* One of the most commonly reported regression metrics.

#### Disadvantages

* Sensitive to outliers.
* Can be heavily influenced by a few extreme prediction errors.

---

### 13.4 R² Score (Coefficient of Determination)

The **R² Score** measures how well the regression model explains the variation in the dependent variable.

The formula is:

$R^2=1-\frac{SS_{Residual}}{SS_{Total}}$  

where:

* $SS_{Residual}$ = Sum of Squared Residuals (unexplained variation)
* $SS_{Total}$ = Total Sum of Squares (total variation in the target variable)

#### Interpretation

| R² Value | Interpretation                                                   |
| -------- | ---------------------------------------------------------------- |
| 0.00     | The model explains none of the variation in the target variable. |
| 0.40     | The model explains 40% of the variation.                         |
| 0.75     | The model explains 75% of the variation.                         |
| 1.00     | Perfect prediction.                                              |

Example:

An R² of **0.82** means that **82% of the variation in the target variable is explained by the predictors**, while the remaining 18% is due to factors not captured by the model or random variation.

#### Advantages

* Easy to interpret as a percentage of explained variance.
* Useful for comparing models built on the same dataset.
* Provides an overall measure of model fit.

#### Disadvantages

* A high R² does not necessarily mean the model is appropriate.
* Adding more predictors can artificially increase R², even if those predictors add little predictive value.
* Does not indicate whether the regression assumptions have been met.

---

### 13.5 Adjusted R²

The **Adjusted R²** improves upon R² by accounting for the number of predictors in the model.

Unlike R², Adjusted R² increases only when a new predictor genuinely improves the model's performance. If an irrelevant predictor is added, Adjusted R² may remain unchanged or even decrease.

This makes it particularly useful when comparing multiple regression models with different numbers of predictors.

#### Why is it important?

Suppose we have two models predicting house prices:

* **Model A:** Size, Bedrooms
* **Model B:** Size, Bedrooms, House Colour

House colour is unlikely to influence house prices. Although adding it may slightly increase the R² value, it provides no meaningful predictive power.

Adjusted R² helps identify this by penalizing unnecessary predictors.

#### Advantages

* Reduces the risk of overfitting.
* Better than R² when comparing models with different numbers of predictors.
* Encourages selecting only meaningful predictors.

#### Disadvantages

* Slightly more difficult to interpret than R².
* Still should not be used as the sole criterion for model selection.

---

## Choosing the Right Evaluation Metric

Different metrics serve different purposes.

| Metric      | Measures                                   | Units            | Sensitive to Outliers | Best Used When                                                             |
| ----------- | ------------------------------------------ | ---------------- | --------------------- | -------------------------------------------------------------------------- |
| MAE         | Average absolute error                     | Original units   | Low                   | All errors are equally important.                                          |
| MSE         | Average squared error                      | Squared units    | High                  | Large errors should be penalized heavily.                                  |
| RMSE        | Average prediction error                   | Original units   | High                  | Large prediction errors are particularly undesirable.                      |
| R²          | Proportion of variance explained           | Percentage (0–1) | No                    | Assessing overall model fit.                                               |
| Adjusted R² | Explained variance adjusted for predictors | Percentage (0–1) | No                    | Comparing multiple regression models with different numbers of predictors. |

---

## Practical Recommendations

When evaluating a regression model, it is good practice to report several metrics rather than relying on just one.

* Use **MAE** when you want an easily interpretable measure of average prediction error.
* Use **RMSE** when large prediction errors are especially costly or should be penalized more heavily.
* Use **MSE** during model training and optimization, particularly because it is the loss function minimized by Ordinary Least Squares regression.
* Use **R²** to understand how much of the variability in the target variable is explained by the model.
* Use **Adjusted R²** when comparing multiple linear regression models with different numbers of predictor variables.

A well-performing regression model typically has **low MAE**, **low MSE**, **low RMSE**, and **high R² and Adjusted R²**, while also satisfying the assumptions of linear regression.

##### FEATURE SCALING
1___. Standardization / Z-score Scaling___   


Another common preprocessing step in Machine Learning is Feature Scaling.  
Feature scaling transforms numerical features so that they are on
approximately the same scale.  

Example:

   Age              : 18 - 60
   Annual Income    : 20,000 - 5,000,000

Without scaling, algorithms may give more importance to features
with much larger numerical values.


Common Types of Scaling


1. Standardization (Z-score Scaling)

   Formula:

      z = (x - mean) / standard deviation

   Characteristics:
      ✓ Mean becomes approximately 0
      ✓ Standard deviation becomes 1

   Commonly used with:
      - Linear Regression
      - Logistic Regression
      - Support Vector Machines (SVM)
      - PCA


2. Min-Max Scaling (Normalization)

   Formula:

      (x - min) / (max - min)

   Characteristics:

      ✓ Values are scaled between 0 and 1.

   Commonly used with:

      - Neural Networks
      - K-Nearest Neighbors (KNN)
      - K-Means Clustering


Do we need scaling in this project?


Our numerical features are:

  - experience_years (0 - 20)
  - skills_count (1 - 19)
  - certifications (0 - 5)

Since these variables already have relatively similar ranges,
Linear Regression can perform well without scaling.

Nevertheless, scaling is an important preprocessing technique and should always be considered when:  
✓ Numerical features have very different ranges.  
✓ Using distance-based algorithms (KNN, K-Means).  
✓ Using Support Vector Machines.  
✓ Using Neural Networks.  
✓ Using regularized regression models such as Ridge or Lasso.  

```python
X_train_encoded = pd.get_dummies(
    X_train,
    columns=categorical_columns,
    drop_first=True,
    dtype=int
)

X_test_encoded = pd.get_dummies(
    X_test,
    columns=categorical_columns,
    drop_first=True,
    dtype=int
)
```

#### STEP 7: Train the Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(X_train_encoded, y_train)
```

#### STEP 8: Make Predictions

```python
y_pred = model.predict(X_test_encoded)
```

```python
results = pd.DataFrame({
    "Actual Salary": y_test.values,
    "Predicted Salary": y_pred
})

results.head(10)
```

#### STEP 9: Evaluate the Model


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error:", round(mae, 2))
print("Root Mean Squared Error:", round(rmse, 2))
print("R2 Score:", round(r2, 4))
```
