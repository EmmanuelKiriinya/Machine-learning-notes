## Introduction to Machine Learning and Types of Machine Learning.
Machine learning is a branch of Artificial Intelligence that allows computers to learn pattern from data and make predictions or decisions without explicitly being programmed for every rule.    
Instead of writing fixed instructions, we give a computer data, and it learns relationships from the data.    
_Example_    
Instead of manually writing rules to determine house rent in Nairobi, we can train a machine learning model using previous housing data. The model learns how factors such as estate, number of bedrooms, house size, availabiliity of parking and distance to the CBD influence rent prices.

After learning fro past data, the model can predict rent for a new house.

### Types of Machine Learning.  
Machine learning can be grouped into three major types:    
1. Supervised Learning
2. Unsupervised learning
3. Reinforcement learning.   

#### 1. Supervised learning
Supervised learning is a type of machine learning where the model learns from labelled data.   
Labelled data means that the dataset already contains the correct answers.    

_Example_    
If we are predicting the monthly rent, the column Monthly rent is the label because it is the answer we want the model to learn.     

|Estate.  |Bedrooms   |House Size(sqm)|Parking  | Monthly rent  |
|---------|-----------|---------------|---------|---------------|
|Umoja    |2          |80             |No       |25000          |
|Umoja    |2          |80             |Yes      |30000          |
|Karen    |4          |220            |Yes      |250000         |
|Kilimani |3          |150            |Yes      |120000         |      

Here:      
__Features:__ Estate, Bedroooms, House Size, parking
__Label:__ Monthly rent.    

___Common types of supervised learning___    
__A). Regression:__ Used when the target variable is a numeric.
_Examples_    
- Predicting the house rent.   
- Predicting salary  
- Predicting sales revenue    

__B). Classification:__ Classification is used when the target variable is a category.    
_Examples_  
- Predict whether a customer will default: Yes or No.    
- Predicting either an email is spam or not spam     
- Predicting whether a student will pass or fail.   

#### 2. Unsupervised learning.     
Unsupervised learning is used when the dataset is not labelled or does not contain any labelled answer.    
The model tries to discover hidden patterns, groups, or structures in the data by itself.     

Example: Assume we have housing data with estate, rent, house size, and distance to cbd, but you do not give a model a target column to predict.    
The model may group houses into cluster such as:    
- Budget houses
- Middle income houses 
- Luxury houses    

___Common types of unsupervised learning___   
- Clustering: Clustering means grouping similar records.   
- Dimensionality reduction: Reduces may columns into fewer important components while still preserving useful information.   

#### 3.Reinforcement learning.    
Reinforcement learning is a type of machine learning where an agent learns by interacting with an environment.    
The agent takes actions and recieves rewards or penalties. The goal is to learn the best actions that maximise reward over time.

### 1. Linear Regression.  
Linear regression is a supervised machine learning algorithm and statistical method used to model the relationship between a dependent variable(target) and one or more independent(predictor) variables.     
Its primary objectives are to:     
- Predict a continous value.     
- Understand how predictors(independent variables) influence the target(dependent variable)  
- Quantifying relationships between variables.    

_Types of Linear Regression_   
1. Simple linear regression => Models the relationship between a single dependent variable and one independent variable using a straight line.     
2. Multiple linear regression => Models the relationship between a single dependent variable and two or more independent variables.   
3. Polynomial linear regression => Models a non-linear relationship by fitting a curved line throught the data using polynomial terms. e.g $y = {0}_1 {x} + {0_2}x^2 + ... {0_n}x^n$    

#### 1.1 Simple Linear Regression.  
Uses one predictor variable(independent variable).  
_Example_    
Predicting salary using years of experience.    

_Formula_   
$y = mx + c$  

ŷ =  β₁x + β₀

y = predicted value or the dependent variable.   
m or β₁ = Slope or the gradient(represents the expected change in Y for every 1-unit increase in X).   
x = predictor or the independent variable.   
c or β₀ = y-Intercept(this is te value of y when X is equal to zero).      

_Formula for slope_    
${beta_1} = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n}(x_i - \bar{x})^2}$ 

alternative formula:    
$m=\frac{n(\sum xy)-(\sum x)(\sum y)}{n(\sum x^{2})-(\sum x)^{2}}$

_Formula for the intercept_    
$β₀ = ȳ - β₁x̄$  

_Example_  
|Experience(x)|Salary_Kes(y)|
|-------------|-------------|
|1.3          |42500        |
|3.2          |63500        |
|7.0          |81800        |
|9.4          |89500        |


````python
import pandas as pd

df = pd.DataFrame({'experience':[1.3,    3.2,    7.0,    9.4   ],
                   'salary':[42500, 63500, 81800, 89500]})

# Get the squares of x
df["experience_squared"] = df["experience"]**2

# Multiplt x and y
df["xy"] = df["experience"] * df["salary"]

# SLope
slope = ((len(df)*(df["xy"].sum())) - (df["experience"].sum() * df["salary"].sum()))/(((len(df)*df["experience_squared"].sum()) - ((df["experience"].sum())**2)))

# Intercept9
intercept = (df["salary"].mean()) - (slope * (df["experience"].mean()))

print("Slope", slope)
print("Y-Intercept", intercept)
````

#### 1.2 Multiple linear regression.    
Uses two or more predictors.     
_Example_      
Predicting house price using, Size, Bedrooms, age.   

_Formula_   
$Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \dots + \beta_nX_n$    
Y = Is the target or the dependent variable.  
$\beta$ = Coeffients.        

##### Definition of terms in Linear Regression.  
- __The Regression Line(Line of best fit):__ The regression line summarizes the average relationship between X and Y. Each point has a predicted value that lies on the line. The line minimizes prediction errors.    
- __Intercept:__ This is the predicted value of Y when X = 0.     
- __Slope:__ The slope tells us how much Y changes when X increases by one unit.     
_Example_    
Sales(y) = 100 + 8(Advertising)   

Interpretation: Every additional 1 shilling spent on advertising increases sales by 8 units.    

- __Residuals(Errors):__ Residual = Actual value - Predicted value     
$Residual = y - \hat{y}$   



````python
y_pred = (5574.24*3.2) + 40199.58
print("Predicted", y_pred)
print("Actual Salary", 63500)

residual = 63500 - y_pred
print("Error/Residual", residual) 
````
